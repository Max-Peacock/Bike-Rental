# Now, lets move on to the Weather data!

## Again, we will call .info() and check, we can see a Dtype that needs updating, and 2 columns that are entirely null which we can remove! We use axis=1 to specify the column, and inplace=True to keep the original DataFrame.

```python
weather_df.info()
```

```python
weather_df['DATE'] = pd.to_datetime(weather_df['DATE'])
weather_df.drop(['PGTM', 'TSUN'], axis=1, inplace=True)
weather_df
```

## We check for duplicates, no duplicates are found which is good

```python
weather_df.duplicated()
```

### Now, we use .describe(), the column names are quite confusing, lets rename some columns and drop some columns, the columns WDF2, WDF5, WSF2, WSF5 are wind speed and direction, the first column AWND (average wind) will provide enough information, so we will remove these 4 Columns. Also, Snowfall in inches is suffice, we do not need the snow depth as well

```python
weather_df.describe()
```

```python
weather_df.drop(['WDF2', 'WDF5', 'WSF2', 'WSF5', 'SNWD'], axis=1, inplace=True)
weather_df
```

```python
weather_df.rename(columns={'STATION': 'Station', 'NAME': 'Name', 'DATE': 'Date', 'SNOW': 'Snowfall_inch','TAVG': 'Average_Temp', 'PRCP': 'Precipitation_Inch', 'AWND': 'Average_Wind_MPH', 'TAVG': 'Average_Temp_C', 'TMAX': 'Max_Temp_C', 'TMIN': 'Min_Temp_C'}, inplace=True)
weather_df
```

### We will now convert the temperatue from Fahrenheit into Celcius

```python
weather_df['Average_Temp_C'] = (weather_df['Average_Temp_C'] - 32) * 5/9
weather_df['Max_Temp_C'] = (weather_df['Max_Temp_C'] - 32) * 5/9
weather_df['Min_Temp_C'] = (weather_df['Min_Temp_C'] - 32) * 5/9
weather_df
```

## We have cleaned the data using pandas, the Database is set up in PGAdmin4 and we have succesfully connected the Database using SQLAlchemy, we will now create the Dataframes and insert Data into the Database

```python

stations_df = combined_df[['Start Station ID', 'Start Station Name', 'Start Station Latitude', 'Start Station Longitude']].drop_duplicates()
stations_df.columns = ['Station_ID', 'Name', 'Latitude', 'Longitude']

users_df = combined_df[['Bike ID', 'User Type', 'Birth Year', 'Gender']].drop_duplicates()
users_df.columns = ['Bike_ID', 'User_Type', 'Birth_Year', 'Gender'] 

trips_df = combined_df[['Bike ID', 'Start Station ID', 'End Station ID', 'Start Time', 'Stop Time', 'Trip Duration', 'User Type', 'Birth Year', 'Gender']].drop_duplicates()
trips_df.columns = [['Bike_ID', 'Start_Station_ID', 'End_Station_ID', 'Start_Time', 'Stop_Time', 'Trip_Duration_Secs', 'User_Type', 'Birth_Year', 'Gender']]

bikes_df = combined_df[['Bike ID']].drop_duplicates()
bikes_df.columns = ['Bike_ID']  

weather_stations_df = weather_df[['Station', 'Name']].drop_duplicates()
weather_stations_df.columns = ['Weather_Station_ID', 'Name']  

weather_data_df = weather_df[['Station', 'Date', 'Average_Wind_MPH', 'Precipitation_Inch', 'Snowfall_inch', 'Average_Temp_C', 'Max_Temp_C', 'Min_Temp_C']]
weather_data_df.columns = ['Weather_Station_ID', 'Date', 'Average_Wind_MPH', 'Precipitation_Inch', 'Snowfall_Inch', 'Average_Temp_C', 'Max_Temp_C', 'Min_Temp_C']
```

## Now, we will insert the data from these Dataframes into our pgAdmin4 database!

```python
# Insert data into 'stations' table
stations_df.to_sql('stations', engine, if_exists='replace', index=False)

# Insert data into 'trips' table
trips_df.to_sql('trips', engine, if_exists='append', index=False)

# Insert data into 'bikes' table
bikes_df.to_sql('bikes', engine, if_exists='append', index=False)

# Insert data into 'users' table
users_df.to_sql('users', engine, if_exists='append', index=False, method='multi', chunksize=1000)

# Insert data into 'weather_stations' table
weather_stations_df.to_sql('weather_stations', engine, if_exists='append', index=False)

# Insert data into 'weather_data' table
weather_data_df.to_sql('weather_data', engine, if_exists='append', index=False)
```
